```python
import pandas as pd
vit=pd.read_csv("vit_sent.csv")
```

```python
import seaborn as sns
```

```python
vit.info()
```

```python
# Vitamin konusunda olumlu duyguları fazla olan ilk 10 il
vit.groupby("Geolocation").polarity.sum().sort_values(ascending=False).head(10)
```

```python
# Günler bazında duygular
vit.groupby("Date").polarity.mean().sort_values(ascending=False).head(20)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,9))
sns.lineplot(x="Geolocation",y="polarity",data=vit)
plt.xticks(rotation=90)
plt.title("Vitamin Sentiment");
```
